### Let's create a rating's df with average rating and number of ratings

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings=pd.DataFrame(df.groupby('title')['rating'].mean())
```

```python
ratings.head(5)
```

```python
ratings['num of ratings']=pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
moviesat=df.pivot_table(index='user_id',columns='title',values='rating')
```

```python
moviesat.head()
```

```python
ratings.sort_values('num of ratings',ascending=False)
```

```python
ratings.head()
```

```python
starwars_user_ratings=moviesat['Star Wars (1977)']
```

```python
starwars_user_ratings.head(50)
```

```python
contact_user_ratings=moviesat['Contact (1997)']
```

```python
similar_to_starwars=moviesat.corrwith(starwars_user_ratings)
similar_to_contact=moviesat.corrwith(contact_user_ratings)
```

```python
corr_starwars=pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False)
```

```python
corr_starwars=corr_starwars.join(ratings['num of ratings'])
```

```python
corr_starwars.head(10)
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation',ascending=False).head(5)
```
